```python
import numpy
import random as ran
#Zad2

def CreateBinTabZero(k):
    inTab = []
    for i in range(k):
        inTab.append(0)
    return inTab

def CreateBinTabOne(k):
    inTab = []
    for i in range(k):
        inTab.append(1)
    return inTab

def CreateRandBinTab(k):
    inTab = []
    for i in range(k):
        inTab.append(ran.randint(0,1))
    return inTab

#def BinToInt():
#    w = 0
#    ii = 0
#    for i in range(len(A) - 1, 0, -1): 
#        if A[i]==1:
#            w += pow(2,ii)
#        ii += 1
#    if A[0] == 1:
#        w = -w
#    return w
#A = [1, 1, 0, 0, 0]
#print(BinToInt())

def BinToInt(inTab):
    w=0
    ii = len(inTab)-2
    for i in inTab[1:]:
        w = w + i*(2**ii)
        ii=ii-1
    if(inTab[0]):
        w=w*-1
    return w

inTab = CreateRandBinTab(6)
print(inTab)
print(BinToInt(inTab))
```

```python
#Zad6 i Zad7
def quick_sort(array):# for now not use
    if len(array) < 2:
        return array
    else:
        pivot = (array[0])
        less = [i for i in array[1:] if i <= pivot]
        greater = [i for i in array[1:] if i > pivot]
        return quick_sort(less) + [pivot] + quick_sort(greater)

def quick_sort_two(tab, population):
    # czymy tab i population w pary (warto_tab, warto_population)
    combined = list(zip(tab, population))
    
    # Sortujemy pary wzgldem wartoci w tab
    combined.sort(key=lambda x: x[0])
    
    # Rozdzielamy posortowane pary na osobne listy
    tab[:], population[:] = zip(*combined)
    return population

def int_to_bin_array(n, size):
    if n == 0:
        return [0] * size if size > 0 else []
    
    binary_digits = []
    
    while n > 0:
        binary_digits.append(n % 2)
        n //= 2

    binary_digits.reverse()

    # Set the size of the output array
    if len(binary_digits) < size:
        binary_digits = [0] * (size - len(binary_digits)) + binary_digits
    elif len(binary_digits) > size:
        # Truncate if the binary representation is larger than the desired size
        binary_digits = binary_digits[-size:]

    return binary_digits

#Ustala przystosowanie chromosomw Wzgldem Posortowania 
def Sort(population):#not optimal 
    size = len(population[0])
    
    for i in range(len(population)):
        population[i]=BinToIntMinusOut(population[i])
    
    population = quick_sort(population)
    
    for i in range(len(population)):
        population[i]=int_to_bin_array(population[i], size)
    
    return population

def Przystosowanie(population):#not optimal 
    size = len(population[0])
    
    for i in range(len(population)):
        population[i]=BinToIntMinusOut(population[i])
    
    tab = []
    for i in range(len(population)):
        tab.append(G(population[i]))
    population = quick_sort_two(tab, population)
    
    for i in range(len(population)):
        population[i]=int_to_bin_array(population[i], size)
    
    return population

def NewPopulationRulet(population):
    #sort from lower to biger
    #Create RuletList (which give more the same index if index is higher (index = 8 have 8...))
    #Use rulet (NewPop)
    
    #Zakadam e lepsze przystosowanie jest Najwiksze u najwikszego chromosomu
    #Np: Populacja[[13][12][1]]
    #                Szansa
    #Populacja[0]      3
    #Populacja[1]      2
    #Populacja[2]      1

    #Create pola Ruletki 
    n=len(population)
    RuletList = []
    for i in range(len(population)+1):
        RuletList.extend([i-1]*i)
    print(RuletList)

    NewPopulation = []
    for i in range(len(population)):
        ran.shuffle(RuletList)
        NewPopulation.append(population[RuletList[0]])
    
    return NewPopulation

population = CreateStartingPopulation(8,4)#whith out this population will be the same as last executed
population = NewPopulationRulet(Przystosowanie(population))
print(population)
```

```python
#Zad8
def Crosing(population, praw):
    new_population = []
    N = len(population)
    population.reverse()
    
    for i in range(0, N, 2):
        parent1 = population[i]
        parent2 = population[(i+1) % N]
        
        # Zastosowanie krzyowania w zalenoci od `crossover_prob`
        if ran.random() < praw:
            # Punkt krzyowania (dla uproszczenia w poowie dugoci osobnikw)
            crossover_point = len(parent1) // 2
            # Tworzenie dzieci
            child1 = parent1[:crossover_point] + parent2[crossover_point:]
            child2 = parent2[:crossover_point] + parent1[crossover_point:]
            new_population.extend([child1, child2])
        else:
            # Jeli brak krzyowania, dodajemy rodzicw do nowej populacji
            new_population.extend([parent1, parent2])
        
        if(len(new_population) > N):
            break
    
    return new_population

population = CreateStartingPopulation(8,4)#whith out this population will be the same as last executed
for i in range(8):
    population=(Crosing(population, 0.9))
    print(population)
```

```python
#Zad9
def PopulationMutation(population, praw):
    new_population = []
    print(population)
    
    for i in population:
        pers = [] 
        for ii in i:
            if ran.random() < praw:
                if ii == 0:
                    pers.append(1)
                else:
                    pers.append(0)
            else:
                pers.append(ii)
        new_population.append(pers)
    
    print(new_population)
    return new_population


population = CreateStartingPopulation(8,4)#whith out this population will be the same as last executed
for i in range(8):
    population = (PopulationMutation(population, 0.05))
    print(population)
```
